```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load image
image = cv2.imread('fruits.jpg', cv2.IMREAD_GRAYSCALE)

# Show image
plt.imshow(image, cmap='gray')
plt.title('Αρχική εικόνα')
plt.show()


# Calculus of gradients with sobel
sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)  # Οριζόντιος
sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)  # Κάθετος

sobel_combined = np.sqrt(sobel_x**2 + sobel_y**2)
sobel_combined = np.uint8(255 * sobel_combined / np.max(sobel_combined))  # Κανονικοποίηση

# Printing Result
plt.imshow(sobel_combined, cmap='gray')
plt.title('Ανίχνευση ακμών με Sobel')
plt.show()


# Canny edge detector
edges = cv2.Canny(image, threshold1=100, threshold2=200)  # Παράμετροι ανάλογα με την εικόνα

# Εμφάνιση αποτελέσματος
plt.imshow(edges, cmap='gray')
plt.title('Ανίχνευση ακμών με Canny')
plt.show()


blurred = cv2.GaussianBlur(image, (5, 5), 0)
edges_clean = cv2.Canny(blurred, 50, 150)
plt.imshow(edges_clean, cmap='gray')
plt.title('Canny με προεπεξεργασία')
plt.show()
```
