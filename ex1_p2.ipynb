```python
from cffi.backend_ctypes import xrange
from skimage.filters import threshold_otsu
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load image
image = cv2.imread('leaf.jpg', cv2.IMREAD_GRAYSCALE)

# Show image
plt.figure(figsize=(12, 5))

plt.subplot(2, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Αρχική εικόνα')
plt.axis('off')
# Calculus of gradients with sobel
sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=5)  # Οριζόντιος
sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=5)  # Κάθετος

sobel_combined = np.sqrt(sobel_x**2 + sobel_y**2)
sobel_combined = np.uint8(255 * sobel_combined / np.max(sobel_combined))  # Κανονικοποίηση

threshold_sobel = threshold_otsu(sobel_combined)
leaf1 = sobel_combined > threshold_sobel

# Printing Result
plt.subplot(2, 2, 2)
plt.imshow(leaf1, cmap='gray')
plt.title('Ανίχνευση ακμών με Sobel')
plt.axis('off')


edges = cv2.Canny(image, 500, 200)
threshold_canny = threshold_otsu(edges)
leaf2 = edges > threshold_canny

plt.subplot(2, 2, 3)
plt.imshow(leaf2, cmap='gray')
plt.title('Ανίχνευση ακμών με Canny')
plt.axis('off')

laplace = cv2.Laplacian(image, cv2.CV_64F, ksize=5)
threshold_laplace = threshold_otsu(laplace)
leaf3 = laplace > threshold_laplace

plt.subplot(2, 2, 4)
plt.imshow(leaf3, cmap='gray')
plt.title('Ανίχνευση ακμών με Laplacian')
plt.axis('off')


plt.show()
```
