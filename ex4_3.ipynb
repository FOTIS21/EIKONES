```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('lenna-n1.jpg', cv2.IMREAD_GRAYSCALE)

# Εφαρμογή φίλτρου μέσου όρου
mean_filtered = cv2.medianBlur(img, 3)
# mean_filtered = cv2.blur(mean_filtered, (5, 5))
# mean_filtered = cv2.blur(mean_filtered, (5, 5))
# mean_filtered = cv2.blur(mean_filtered, (5, 5))
# mean_filtered = cv2.blur(mean_filtered, (5, 5))


# Εμφάνιση
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title("Αρχική (με θόρυβο)")
plt.imshow(img, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title("Mean Filter (5x5)")
plt.imshow(mean_filtered, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```
