```python
from matplotlib import pyplot as plt
import pytesseract
from PIL import Image
import cv2

image = cv2.imread('book-cover.jpeg', cv2.IMREAD_GRAYSCALE)
text = pytesseract.image_to_string('book-cover.jpeg', lang='eng')

print(text)

plt.imshow(image, cmap='gray')
plt.axis('off')
plt.show()
```
